```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook
from bokeh.sampledata.iris import flowers

from bokeh.transform import factor_cmap
```

```python
import numpy  as np
```

```python
output_notebook()
```

```python
## Creando la figura
data = ColumnDataSource(flowers)
print(data.data.keys())
```

```python
f = figure(plot_width = 800, plot_height = 400)

SPECIES = np.unique(data.data['species'])
# Colormap
col_map = factor_cmap('species', 'Set1_3', SPECIES)

# Se crea el plot
f.scatter(x = 'petal_width', y = 'petal_length', 
          size = 'sepal_width', source = data,
          color = col_map, 
          line_color = col_map,
          legend = 'species')
# TÃ­tulo
f.title.text = 'Iris Morphology'
f.title.text_font = 'times'
f.title.align = 'center'
f.title.text_font_size = '1.2em'
# Labels
f.yaxis.axis_label = 'Petal Width'
f.xaxis.axis_label = 'Petal Length'
f.axis.axis_label_text_font_style = 'bold'
f.axis.axis_label_text_font = 'times'

show(f)
```
